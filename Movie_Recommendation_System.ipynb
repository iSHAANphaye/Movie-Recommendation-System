```python
import ast # Abstract Syntax Tree
```

```python
def convert(obj):      # Will help in converting the complex syntax into easy-to-read ones, separated by commas, containing only the 'name' parameters.
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)  # Applying the 'convert' function
movies_df['keywords']=movies_df['keywords'].apply(convert)
movies_df['production_companies']=movies_df['production_companies'].apply(convert)
```

```python
movies_df.head()
```

```python
def fetch_actors(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies_df['cast']=movies_df['cast'].apply(fetch_actors)
```

```python
movies_df.head()
```

```python
def fetch_director(obj):    # Fetch the director's name from crew column
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch_director)
```

```python
from sklearn.feature_extraction.text import CountVectorizer  #  It is used to transform a given text into a vector on the basis of the frequency (count) 
cv = CountVectorizer(max_features=5000,stop_words='english')  #  of each word that occurs in the entire text.
```

1. The "fit(data)" method is used to compute the mean and std dev for a given feature to be used further for scaling.
2. The "transform(data)" method is used to perform scaling using mean and std dev calculated using the .fit() method.
3. The "fit_transform()" method does both fits and transform.

```python
cv.fit_transform(new_df['tags']).toarray().shape    # Creates a dictionary of tokens, and converts texts to numbers
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()   # Convert vector to array
```

STEMMING and LEMMATIZATION are Text Normalization (or sometimes called Word Normalization) techniques in the field of Natural Language Processing that are used to prepare text, words, and documents for further processing.

STEMMING is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in the Language. Ex. played, playing and plays all 3 have the root word "play". Hence, stemming is done to obtain "play"

```python
from nltk.stem.porter import PorterStemmer   # nltk = natural language tool-kit
ps = PorterStemmer()
```

```python
def stemm(text):    # function for applying stemming to every word in the string, and return the stemmed string
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stemm)   # Apply stemming function to the 'tags' list in new_df
```

Cosine similarity measures the similarity between two vectors of an inner product space. It is measured by the cosine of the angle between two vectors and determines whether two vectors are pointing in roughly the same direction

```python
from sklearn.metrics.pairwise import cosine_similarity     
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

The code below will return the element in that array whose second element (x[1]) is larger than all of the other elements' second elements. Another way of phrasing it is as the function call implies: return the max element, using x[1] as the key. Also, reverse is True, hence the list will be reversed.

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):                    
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
